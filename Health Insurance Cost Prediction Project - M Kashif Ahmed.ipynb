```python
# checking the outliers count by IQR approach

Q1 = df['charges'].quantile(0.25)
Q3 = df['charges'].quantile(0.75)
IQR = Q3 - Q1

# Defining the upper and lower bounds for outliers
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

# Counting the outliers
# anything which is greater than upper bound will detect as outliers & anything which is lesser then lower bound is detected as outliers
outliers_count = ((df['charges'] > upper_bound) | (df['charges'] < lower_bound)).sum()

#print(f'The number of outliers is: {outliers_count}')
print('The number of outliers count is:',outliers_count)
```

```python
# Train & Test Split Using R2--R Square Approach we will find the r2 score
```

```python
# importing the necessary libraries for prediction of the model

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.metrics import r2_score
import numpy as np
```

```python
x=df.drop('charges',axis=1) # here x in independent column
x
```

```python
y=df['charges'] # here y is dependent column
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
prediction=model.predict(x_train)
```

```python
prediction
```

```python
r2=metrics.r2_score(y_train,prediction)
```

```python
print('The R Square is :',r2) # the r square value is around 0.75
```
